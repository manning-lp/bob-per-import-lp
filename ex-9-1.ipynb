```python
# 3b. using import test_module to load test_module and observe output... why?
import test_module

# received 'counter is 1' because import loads and runs the contents of the module when it is import'ed
```

```python
# 3c. use dir() and sys.modules to confirm test_module IS loaded

print(f'{dir()=}')

import sys
print(f'{sys.modules=}')

if 'test_module' in dir():
    print('found test_module in dir()')

if 'test_module' in sys.modules:
    print('found test_module in sys.modules')
```

```python
# 3d. use the __doc__ property to get the module docstring and the id() function to get its ID

print(f'{test_module.__doc__=}')
print(f'{id(test_module)=}')
```

```python
# 3e. import test_module again and see if ID changes and if output prints second time

import test_module

print(f'{id(test_module)=}')

# output did NOT print second time, and ID is same
```

```python
# 4a. RESTART KERNEL, then import just the return_source_module function ... same output? what means?

from test_module import return_source_module

# received counter is 1 again ... means the module level statements are executed even when a single function is import'ed
```

```python
# 4b. Use dir() to see if return_source_module is available

print(f'{dir()=}')

if 'return_source_module' in dir():
    print('found return_source_module in dir()')
```

```python
# 4c. use sys.modules to see if test_module is there, if yes, assign and use dir() to see contents

import sys # because restarted kernel

print(f'{sys.modules=}')
print(f'{sys.modules['test_module']=}')
test_module_var = sys.modules['test_module']
print(f'{dir(test_module_var)=}')
```

```python
# 4d. use id() to get the ID of the module
print(f'{id(test_module_var)=}')
```

```python
# 4e. import the whole module with out reset kernel.Output? check dir() and id() again
import test_module

print(f'{dir()=}')
print(f'{id(test_module)=}')

# in dir() and ID same as above
```

```python
# 4f. experiment with deleting test_module module and deleteing from sys.modules...

del test_module
del sys.modules['test_module']

print(f'{dir()=}')
if 'return_source_module' in dir():
    print('found return_source_module in dir()')
if 'test_module' in dir():
    print('found test_module in dir()')
if 'test_module' in sys.modules:
    print('found test_module in sys.modules')

# return_source_module is still there, but test_module is not
```
